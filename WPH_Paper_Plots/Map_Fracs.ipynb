```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
sq_deg_in_sky = 4*np.pi*(180/np.pi)**2
```

```python
mask_dir = '/n/netscratch/dvorkin_lab/Lab/kboone/cmb_conv/masks/'
cmbl_dir = '/n/netscratch/dvorkin_lab/Lab/gvalogiannis/maps_cmbl_georgios/maps_cmbl_georgios/'
```

```python
surveys = ['ACT', 'Planck', 'SO', 'SPT']
```

```python
cmbl = np.load(cmbl_dir + 's8_99.npy', allow_pickle=True)[0]
cmbl_masks = []

for survey in surveys:
    if (survey == 'Planck') or (survey == 'SPT'):
        cmbl_masks.append(np.load(mask_dir + survey + '.npy'))
    else:
        ind_mask = np.zeros(12*1024**2)
        ind_mask[np.where(cmbl[survey] != 0)[0]] = 1
        cmbl_masks.append(ind_mask)
        
wl_mask = np.load('/n/netscratch/dvorkin_lab/Lab/kboone/cmb_x_wl/wl_masks/dr3.npy')
```

```python
survey_ind = 3
print(surveys[survey_ind])
print(np.sum(cmbl_masks[survey_ind]) / (12*1024**2) * sq_deg_in_sky)
```

```python
print(np.sum((wl_mask == cmbl_masks[survey_ind]) & (wl_mask == 1)) / (12*1024**2) * sq_deg_in_sky)
```
