```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nz = np.loadtxt('../Data/nofz.txt')

redshift = nz[:,0]
bin_1 = nz[:,1]+nz[:,2]
bin_2 = nz[:,3]+nz[:,4]
bin_3 = nz[:,5]+nz[:,6]
bin_4 = nz[:,7]+nz[:,8]
bin_5 = nz[:,9]+nz[:,10]
bin_6 = nz[:,11]+nz[:,12]
```

```python
plt.figure(dpi = 120)
plt.hist(redshift, weights=bin_1, alpha = 0.5, bins = redshift, label = 'Low z Bin', density=True)
# plt.hist(redshift, weights=bin_2, alpha = 0.5, bins = redshift)
# plt.hist(redshift, weights=bin_3, alpha = 0.5, bins = redshift)
# plt.hist(redshift, weights=bin_4, alpha = 0.5, bins = redshift)
# plt.hist(redshift, weights=bin_5, alpha = 0.5, bins = redshift)
plt.hist(redshift, weights=bin_6, alpha = 0.5, bins = redshift, label = 'High z Bin', density=True)
plt.legend()
plt.grid()
plt.ylabel('Normalized N(z)')
plt.xlabel('z')
plt.xlim([0,2])
plt.title('Galaxy Redshift Distributions in Tomographic Bins')
plt.show()
```
