```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
from Bin_Helpers import *
import CompressedFisher
import numpy as np
import matplotlib
```

```python
matplotlib.style.use('des_dr1')
```

```python
numBins = 15
data_len = 2000
pivots = [[200, 0.3*20], [1250, 45*0.3]]
Qp = create_linlin_matrix(data_len, numBins, points=pivots)
Q = create_linlin_matrix(data_len, numBins)
```

```python
fig = plt.figure(figsize=(10, 5), dpi=120)
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1], wspace=0.05)

# Plot Qp.T with annotations
ax0 = plt.subplot(gs[0])
im = ax0.imshow(Qp.T, aspect='auto', cmap='Greys')
ax0.set_title('Binning Matrix with Two Pivots')
ax0.set_ylabel('Compressed Index')
ax0.set_xlabel('Original Index')

# Annotations
ax0.annotate(
    'Pivots',
    xy=pivots[0],
    xytext=(850, 21.5*0.3),
    arrowprops=dict(arrowstyle='->', color='k', shrinkB=5),
    fontsize=12
)
ax0.annotate(
    'Pivots',
    xy=pivots[1],
    xytext=(850, 21.5*0.3),
    arrowprops=dict(arrowstyle='->', color='k', shrinkB=5),
    fontsize=12
)

# Plot Q without annotations
ax1 = plt.subplot(gs[1], sharey=ax0)
im2 = ax1.imshow(Q.T, aspect='auto', cmap='Greys')
ax1.set_title('Binning Matrix without Pivots')
ax1.set_xlabel('Original Index')
plt.setp(ax1.get_yticklabels(), visible=False)  # Hide redundant y-tick labels

# Shared colorbar
cbar = fig.colorbar(im, ax=[ax0, ax1], orientation='vertical', fraction=0.02, pad=0.02)
cbar.set_label('Value')

# plt.savefig('../Paper_Plots/Binning_Matrices.pdf', bbox_inches='tight', format='pdf')
# plt.close()
plt.show()
```
