```python
%matplotlib inline
```

```python
Om = 0.3175
s8 = 0.834
deriv_params = ['Om', 's8']
```

```python
surveys = ['ACT', 'Planck', 'SPT', 'SO']
pers = [25, 50, 75, 100]

Oms = []
s8s = []

for s in surveys:
    survey_Oms = []
    survey_s8s = []
    for per in pers:
        covs = np.load('Splits_Data/Cross/Cl_'+str(per)+'_'+s+'_x_dr3_contours.npy')
        cov = np.average(covs, axis=0)
        survey_Oms.append(np.sqrt(cov[0,0]))
        survey_s8s.append(np.sqrt(cov[1,1]))
    
    Oms.append(survey_Oms)
    s8s.append(survey_s8s)

Oms = np.array(Oms)
s8s = np.array(s8s)
```

```python
plt.figure(dpi = 120)
plt.axhline(y=1, color='k', linestyle='--')
for i in np.arange(len(surveys)):
    plt.plot(pers, Oms[i]/Oms[i,-1], label=surveys[i] + r' $\times$ Euclid DR2', marker='o')
plt.title(r'Cross $C_\ell$ Constraint Convergence for $\Omega_m$', fontsize=24)
plt.xlabel('Number of Derivatives Used', fontsize=24)
plt.ylabel(r'$\sigma(\Omega_m)$ / Final Constraint', fontsize=24)
plt.legend()
# plt.grid()
# plt.show()
plt.savefig('../Paper_Plots/CRO_Om_Conv.pdf', format='pdf', bbox_inches='tight',  # include everything that sticks out
              pad_inches=0.02)
plt.close()
```

```python
plt.figure(dpi = 120)
plt.axhline(y=1, color='k', linestyle='--')
for i in np.arange(len(surveys)):
    plt.plot(pers, s8s[i]/s8s[i,-1], label=surveys[i] + r' $\times$ Euclid DR2', marker='o')
plt.title(r'Cross $C_\ell$ Constraint Convergence for $\sigma_8$', fontsize=24)
plt.xlabel('Number of Derivatives Used', fontsize=24)
plt.ylabel(r'$\sigma(\sigma_8)$ / Final Constraint', fontsize=24)
plt.legend()
# plt.grid()
# plt.show()
plt.savefig('../Paper_Plots/CRO_s8_Conv.pdf', format='pdf', bbox_inches='tight',  # include everything that sticks out
              pad_inches=0.02)
plt.close()
```
